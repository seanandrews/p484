```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.colors as col
import sys
sys.path.append('/data/sandrews/jhfiles/LPfiles/figures')
from continuumanalysis3 import Continuum

```

```python
cont = Continuum('/pool/firebolt0/LPscratch_new/Elias27/Elias27_SB_contap.image.fits', 0.9*.03,-0.9*.03, 117.3, 55.8, 120)
radialbins = 2.5+5*np.arange(60)
surfbrightness, SB_scatter, rtmap = cont.azunwrap(radialbins, tbins = -175 + 10.*np.arange(36))
#plt.imshow(rtmap,extent =[0, np.max(radialbins)+2.5, -180,180], aspect = 0.4, origin = 'lower', cmap = 'cubehelix',norm = col.PowerNorm(gamma = 0.5), vmin = 0)
NWmodel = np.zeros((370,2)) #columns are theta and r
NWmodel[:,0] = -np.arange(100,470)
NWmodel[:,1] = 84*120/139.*np.exp(-0.138*NWmodel[:,0]*np.pi/180.) #negative sign on b is because the angles are decreasing with radius


SEmodelpart1 = np.zeros((200,2))
SEmodelpart1[:,0] = -np.arange(200,400)
SEmodelpart1[:,1] = 84*120/139.*np.exp(-0.138*SEmodelpart1[:,0]*np.pi/180.)
SEmodelpart2 = np.zeros((200,2))
SEmodelpart2[:,0] = -np.arange(200,400)
SEmodelpart2[:,1] = 84*120/139.*np.exp(-0.138*SEmodelpart1[:,0]*np.pi/180.)
plt.imshow(rtmap*radialbins**2,extent =[0, np.max(radialbins)+2.5, -180,180], aspect = 0.65, origin = 'lower', cmap = 'magma', vmin = 0, interpolation = 'bicubic')
plt.scatter(NWmodel[:,1], NWmodel[:,0]+270, color = 'white', s = 1)
plt.scatter(SEmodelpart1[:,1], SEmodelpart1[:,0]+90, color = 'white', s = 1)
plt.scatter(SEmodelpart2[:,1], SEmodelpart2[:,0]+450, color = 'white', s = 1)
plt.ylim(ymin = -180, ymax = 180)
plt.xlabel('Radius (AU)')
plt.ylabel('Theta (degrees)')
```

```python
plt.imshow((rtmap-np.percentile(rtmap, 25, axis = 0))*radialbins**2,extent =[0, np.max(radialbins)+2.5, -180,180], aspect = 0.65, origin = 'lower', 
           cmap = 'magma', interpolation = 'bicubic')
plt.scatter(NWmodel[:,1], NWmodel[:,0]+270, color = 'white', s = 1)
plt.scatter(SEmodelpart1[:,1], SEmodelpart1[:,0]+90, color = 'white', s = 1)
plt.scatter(SEmodelpart2[:,1], SEmodelpart2[:,0]+450, color = 'white', s = 1)
plt.ylim(ymin = -180, ymax = 180)
plt.xlabel('Radius (AU)')
plt.ylabel('Theta (degrees)')
```

```python
fig = plt.figure(figsize = (7,7))
circularized = cont.circularize()
plt.imshow(circularized[600:-600, 300:-300].T, origin = 'lower',norm = col.PowerNorm(gamma = 0.5), vmin = 0, cmap = 'jet', vmax = 0.02)

```

```python
plt.semilogy(radialbins,np.mean(rtmap, axis = 0),label = 'mean')
plt.semilogy(radialbins, np.percentile(rtmap, 50, axis = 0), label = 'median')
plt.semilogy(radialbins, np.percentile(rtmap, 25, axis = 0), label = 'first quartile')
plt.semilogy(radialbins, np.percentile(rtmap, 10, axis = 0), label = 'first decile')
plt.semilogy(radialbins, np.min(rtmap, axis = 0), label = 'minimum')
plt.legend()
```
